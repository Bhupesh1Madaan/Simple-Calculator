```python
import tkinter as tk

def calculate():
    try:
        a = int(entry1.get())
        b = int(entry3.get())
        operator = entry2.get()

        if operator == "+":
            result.set(a + b)
        elif operator == "-":
            result.set(a - b)
        elif operator == "*":
            result.set(a * b)
        elif operator == "/":
            result.set(a / b)
        elif operator == "%":
            result.set(a % b)
        else:
            result.set("Invalid Operator")
    except ValueError:
        result.set("Invalid Input")

def clear_entries():
    entry1.delete(0, tk.END)
    entry2.delete(0, tk.END)
    entry3.delete(0, tk.END)
    result.set("")

def on_exit():
    root.destroy()

root = tk.Tk()
root.title("Simple Calculator")

# Styling
root.geometry("300x200")
root.resizable(True,True)
root.configure(bg="#f0f0f0")

# Create entry widgets
entry1= ("Enter Number:")
entry1 = tk.Entry(root, width=10, font=('Helvetica', 12))
# entry2=input("Enter Operator:")
entry2 = tk.Entry(root, width=5, font=('Helvetica', 12))
# entry3=input("Enter Number:")
entry3 = tk.Entry(root, width=10, font=('Helvetica', 12))

# Create result variable
result = tk.StringVar()

# Create label to display result
result_label = tk.Label(root, textvariable=result, font=('Helvetica', 14), bg="#f0f0f0")

# Create buttons and attach functions
calculate_button = tk.Button(root, text="Calculate", command=calculate, font=('Helvetica', 12), bg="#4caf50", fg="white")
clear_button = tk.Button(root, text="Clear", command=clear_entries, font=('Helvetica', 12), bg="#2196f3", fg="white")
exit_button = tk.Button(root, text="Exit", command=on_exit, font=('Helvetica', 12), bg="#f44336", fg="white")

# Arrange widgets in the grid
entry1.grid(row=0, column=3, padx=10, pady=10)
entry2.grid(row=1, column=3, padx=15, pady=10)
entry3.grid(row=2, column=3, padx=10, pady=10)
result_label.grid(row=1, column=1, columnspan=3, pady=10)
calculate_button.grid(row=3, column=3, pady=10, padx=5, sticky="ew")
clear_button.grid(row=3, column=2, pady=10, padx=5, sticky="ew")
exit_button.grid(row=3, column=1, pady=10, padx=5, sticky="ew")

# Set column weights to make buttons expand with window resizing
root.columnconfigure(0, weight=1)
root.columnconfigure(1, weight=1)
root.columnconfigure(2, weight=1)

# Start the Tkinter event loop
root.mainloop()
```
